```python
#9.5.cuvinte ce contin doar litere cerute

fin = open('words.txt')

def without_s(z, s):
    y = ""
    for letter in z:
        if letter not in s:
            y += letter
    return y


def uses_all(word, s, z):
    q = without_s(z, s)
    for i in word:
        if i in q:
            return False
    for i in s:
        if i not in word:
            return False 
    return True
   

x = raw_input("Introduceti va rog caracterele pe care doriti sa le folositi: ")

z = 'qwertyuiopasdfghjklzxcvbnm' #alfabetul
for line in fin:
    word = line.strip()
    if uses_all(word, x, z):
        print word

```

```python
#9.6

fin = open('words.txt')

def is_abecedarian(word):
    if len(word) <= 1:
        return True
    if word[0] > word[1]:
        return False
    return is_abecedarian(word[1:])
        
k = 0
for line in fin:
    word = line.strip()
    if is_abecedarian(word):
        k += 1
print 'Exista ', k, ' cuvinte care sunt formate din litere in ordine lexicografica'
        
```
